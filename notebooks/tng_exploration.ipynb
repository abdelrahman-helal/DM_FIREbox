```python
# Load the data
halos = pd.read_parquet('../data/tng-data/TNG300-1-halos_99.parquet')
subhalos = pd.read_parquet('../data/tng-data/TNG300-1-subhalos_99.parquet')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(subhalos['subhalo_logstellarmass'], subhalos['subhalo_loghalomass'], 
            alpha=0.5, c=subhalos['subhalo_logstellarhalfmassradius'], 
            cmap='plasma', s=10)
plt.colorbar(label='Subhalo Stellar Half Mass Radius (log)')
plt.xlabel('Stellar Mass (log)')
plt.ylabel('DM Halo Mass (log)')
plt.show()
```
